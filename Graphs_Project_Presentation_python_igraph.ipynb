```python
import igraph as ig
import matplotlib.pyplot as plt
```

# Tworzenie grafu od podstaw

```python
# n - liczba wierzchołków, musi być liczbą naturalną
n = 8
# edges - lista krawędzi, musi być listą dwuelementowych krotek
edges = []

# Tworzenie grafu - pierwszy argument: liczba wierzchołków, drugi argument: lista krawędzi
graf = ig.Graph(n, edges)

# Tytuł grafu
graf["title"] = "Graf bez krawędzi" # Tytuł grafu

# Przypisywanie wierzcchołkom - g.vs["cecha"] = ["wariant cechy 1", "wariant cechy 2", "wariant cechy 3"]
graf.vs["nazwa"] = ["v1", "v2", "v3", "v4", "v5", "v6", "v7", "v8"]
graf.vs["indeks"] = [1, 2, 3, 4, 5, 6, 7, 8]
# Wartości liczbowe również są dopuszczalne jako etykiety


fig, ax = plt.subplots(figsize=(7,7))

# Funkcja pakietu igraph do drukowania utworzonego grafu
ig.plot(
    graf,
    target=ax,
    #layout="circle",               # wymuszenie kolistego układu wierzchołków
    #vertex_label=graf.vs["indeks"] # wyświetlenie etykiet wierzchołków, używając jako etykiety określonej cechy
    vertex_label=graf.vs["nazwa"],  # wyświetlenie etykiet wierzchołków, używając jako etykiety określonej cechy
    vertex_label_size = 30           # rozmiar wyświetlanej etykiety wierzchołka
)
plt.show()
```

## Tworzenie krawędzi - krawędzie wielokrotne

```python
# n - liczba wierzchołków, musi być liczbą całkowitą
n = 5
# edges - lista krawędzi, musi być listą dwuelementowych krotek
# By stworzyć pętlę, krawędź wielokrotną, należy kilkukrotnie zamieścić w liście krawędzi tę samą krotkę
# Wewnątrz programu, wierzchołki numerowane są od zera, więc podanie krotki zawierającej liczbę a większą lub równą n spowoduje dodanie nowego wierzchołka.
edges = [(0, 1), (0, 1), (0, 1)]

# Tworzenie grafu - pierwszy argument: liczba wierzchołków, drugi argument: lista krawędzi
graf = ig.Graph(n, edges)

# Tytuł grafu
graf["title"] = "Graf bez krawędzi" # Tytuł grafu

# Przypisywanie wierzcchołkom - g.vs["cecha"] = ["wariant cechy 1", "wariant cechy 2", "wariant cechy 3"]
graf.vs["nazwa"] = ["v1", "v2", "v3", "v4", "v5", "v6", "v7", "v8"]
graf.vs["indeks"] = [1, 2, 3, 4, 5, 6, 7, 8]
# 


fig, ax = plt.subplots(figsize=(8,8))

# Funkcja pakietu igraph do drukowania utworzonego grafu
ig.plot(
    graf,
    target=ax,
    #layout="circle",               # wymuszenie kolistego układu wierzchołków
    #vertex_label=graf.vs["indeks"] # wyświetlenie etykiet wierzchołków, używając jako etykiety określonej cechy
    vertex_label=graf.vs["nazwa"],  # wyświetlenie etykiet wierzchołków, używając jako etykiety określonej cechy
    #vertex_label_size=30           # rozmiar wyświetlanej etykiety wierzchołka
)
plt.show()
```

##Drukowanie grafu, dodawanie i usuwanie krawędzi oraz wierzchołków

Dodawanie wierzchołków i krawędzi

```python
#Dodać pojedynczy wierzchołek możemy metodą add_vertex("etykieta wierzchołka"), dodać n wierzchołków naraz możemy metodą add_vertices
graf.add_vertex()
print(graf)

# Dodać n wierzchołków naraz można metodą .add_vertices(n)
#graf.add_vertices(5)

#Dodać pojedynczą krawędź między wierzchołkami v_1 i v_2 można metodą add_edge(v_1, v_2)
graf.add_edge(2, 3)

graf.vs["nazwa"] = ["v1", "v2", "v3", "v4", "v5", "v6", "v7", "v8"]
graf.vs["indeks"] = [1, 2, 3, 4, 5, 6, 7, 8]

fig, ax = plt.subplots(figsize=(8,8))

# Funkcja pakietu igraph do drukowania utworzonego grafu
ig.plot(
    graf,
    target=ax,
    #layout="circle",               # wymuszenie kolistego układu wierzchołków
    #vertex_label=graf.vs["indeks"] # wyświetlenie etykiet wierzchołków, używając jako etykiety określonej cechy
    vertex_label=graf.vs["nazwa"],  # wyświetlenie etykiet wierzchołków, używając jako etykiety określonej cechy
    #vertex_label_size=30           # rozmiar wyświetlanej etykiety wierzchołka
)
plt.show()

# Dodane wierzchołki będą bezimienne, jeżeli program nie ma do dyspozycji listy etykiet w tej samej komórce.
```
