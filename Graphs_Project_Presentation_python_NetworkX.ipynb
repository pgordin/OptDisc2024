# Tworzenie grafu od podstaw

```python
import matplotlib.pyplot as plt
# Utworzenie pustego grafu

graf = nx.Graph()

# Dodawanie pojedynczego wierzchołka
# metoda .add_node() bierze jeden argument, którym jest etykieta nowego wierzchołka- etykieta może byc liczbą, tekstem lub krotką, ale nie listą

graf.add_node(1)

subax1 = plt.subplot(111)

nx.draw(graf, with_labels=True, connectionstyle='arc3, rad = 0.1', font_weight='bold') # Utworzenie rysunku grafu
plt.show()
```

```python
# dodawanie wierzchołków z listy
# By dodać kilka wierzchołków naraz,
# skorzystać trzeba z metody .add_nodes_from(). Metoda ta bierze jeden argument: listę etykiet nowych wierzchołków
graf.add_nodes_from([2, 7])

subax1 = plt.subplot()

nx.draw(graf, with_labels=True, connectionstyle='arc3, rad = 0.1', font_weight='bold') # Utworzenie rysunku grafu
plt.show()
```

```python
# Dodanie pojedynczej krawędzi możliwe jest za pomocą metody .add_edge()
# graf.(v_i,v_j), gdzie v_i i v_j to wierzchołki między którymi chcemy utworzyć krawędź
graf.add_edge(1, 7)
# Podanie wierzchołka nieistniejącego powoduje utworzenie tego wierzchołka w grafie
# Pętlę tworzy się podając dwa razy ten sam wierzchołek
subax1 = plt.subplot()

nx.draw(graf, with_labels=True, connectionstyle='arc3, rad = 0.1', font_weight='bold') # Utworzenie rysunku grafu
plt.show()
```

```python
# Dodanie kilku krawędzi naraz możliwe jest za pomocą metody
# add_edges_from([(v_i, v_j), (v_n, v_m)]),
# gdzie argumentem jest lista par wierzchołków określających końce krawędzi, które mają być dodane
graf.add_edges_from([(2, 1), (2, 7)])

subax1 = plt.subplot()

nx.draw(graf, with_labels=True, connectionstyle='arc3, rad = 0.1', font_weight='bold') # Utworzenie rysunku grafu
plt.show()
```

## Etykietowanie wierzchołków - przypisywanie atrybutów

```python
# By nadać wierzchołkowi atrybut podczas dodawania, trzeba skorzystać z metody
# .add_nodes_from()
# gdzie argumentem jest lista dwuelementowych krotek o budowie:
# (etykieta wierzchołka, słownik złożony z par: "kategoria":"cecha")

graf.add_nodes_from([(3, {"President":"Trump"}), (4, {"President":"Reagan"})])

subax1 = plt.subplot()

nx.draw(graf, with_labels=True, connectionstyle='arc3, rad = 0.1', font_weight='bold') # Utworzenie rysunku grafu
plt.show()
```

## Etykietowanie krawędzi - przypisywanie atrybutów

```python
# By nadać krawędzi atrybut (np. wagę) podczas jej dodawania, trzeba skorzystać z metody
# .add_edges_from()
# gdzie argumentem jest lista trójementowych krotek o budowie:
# (v_i, v_j, słownik złożony z par: "kategoria":"cecha")

graf.add_edges_from([(3, 2, {"weight":5}), (4, 3, {"weight":13})])

subax1 = plt.subplot()

nx.draw(graf, with_labels=True, font_weight='bold') # Utworzenie rysunku grafu
plt.show()
```

## Krawędzie wielokrotne i pętle

### Pętle

```python
# By dodać pętlę do utworzonego grafu, wystarczy skorzystać z metody
# .add_edge(v_i, v_i)
# i podać jej za obydwa argumenty ten sam wierzchołek
graf.add_edge(1,1)

subax1 = plt.subplot()
nx.draw(graf, with_labels=True, connectionstyle='arc3, rad = 0.1', font_weight='bold') # Utworzenie rysunku grafu
plt.show()
print(graf)
```

### Krawędzie wielokrotne

```python
# Brak porstego rozwiązania
graf.add_edge(1,7)

subax1 = plt.subplot()
pos = nx.spring_layout(graf)
nx.draw(graf, pos, with_labels=True, connectionstyle='arc3, rad = 0.1') # Utworzenie rysunku grafu
plt.show()
print(graf)
```

# Generowanie grafów

## Graf losowy $G_{n,p}$

```python
# By wygenerować losowy graf, gdzie
# n - liczba wierzchołków
# p - prawdopowodobieństwo pojawienia się krawędzi między wierzchołkami, 0 <= p <= 1
# posłużyć się trzeba funkcją
# gnp_random_graph(n, p)
rand_graph = nx.gnp_random_graph(5, 0.25)

subax1 = plt.subplot()
nx.draw(rand_graph, with_labels=True, font_weight='bold') # Utworzenie rysunku grafu
plt.show()
```

## Graf losowy $G_{n,m}$

```python
# By wygenerować graf, gdzie
# n - liczba wierzchołków
# m - liczba krawędzi,
# przy czym rozmieszczenie krawędzi jest losowe, posłużyć się trzeba funkcją
# gnm_random_graph(n, m)

n_m_graph = nx.gnm_random_graph(4, 3)
# UWAGA: wygenerowany graf będzie grafem prosty, co najwyżej grafem pełnym

subax1 = plt.subplot()
nx.draw(n_m_graph, with_labels=True, font_weight='bold') # Utworzenie rysunku grafu
plt.show()
print(n_m_graph)
```

## Graf regularny

```python
# By wygenerować graf regularny, czyli taki, w którym wszystkie wierzchołki są tego samego stopnia,
# posłużyć się trzeba funkcją
# .random_regular_graph(d, n)
# d - stopień wierzchołków
# n - liczba wierzchołków
# przy czym n * d musi być liczbą parzystą oraz d < n

reg_graph = nx.random_regular_graph(2,4)

subax1 = plt.subplot()
nx.draw(reg_graph, with_labels=True, font_weight='bold') # Utworzenie rysunku grafu
plt.show()
print(reg_graph)
```

## Losowe drzewo

```python
# By wygenerować losowe drzewo o n wierzchołkach, posłużyć się trzeba funkcją
# random_tree()

tree = nx.random_tree(8)

subax1 = plt.subplot()
nx.draw(tree, with_labels=True, font_weight='bold') # Utworzenie rysunku grafu
plt.show()
print(tree)
```

## Graf na podstawie listy stopni wierzchołków
Mając n-elementową listę stopni oczekiwanych, program tworzy graf o n wierzchołkach i prawdpodobieństwie wystąpienia krawędzi obliczonym na podstawie
podanej listy.

```python
# By wygenerować na podstawie listy stopni wierzchołków, posłużyć się trzeba funkcją
# configuration_model(lista_stopni_wierzchołków)
# Suma stopni wierzchołków musi być parzysta
list_graph = nx.configuration_model([0, 0, 2, 2])

subax1 = plt.subplot()
nx.draw(list_graph, with_labels=True, font_weight='bold') # Utworzenie rysunku grafu
plt.show()
print(list_graph)
# Program ma problemy z rysowaniem krawędzi wielokrotnych
```

## Atlas grafów
wszystkie grafy proste spójne złożone z do 6 wierzchołków

```python
# "Atlas" zawiera 142 grafy. By przywołać dowolny graf, należy użyć funkcji
# graph_atlas(i)
# i - liczba od 1 do 142

atlas_graph = nx.graph_atlas(45)

subax1 = plt.subplot()
nx.draw(atlas_graph, with_labels=True, font_weight='bold') # Utworzenie rysunku grafu
plt.show()
print(atlas_graph)
```

## Przeszukiwanie BFS i DFS

```python
subax1 = plt.subplot()
nx.draw(tree, with_labels=True, font_weight='bold') # Utworzenie rysunku grafu
plt.show()
print(tree)
```

### DFS

```python
# By zastosować algorytm DFS, skorzystać należ z funkcji
# list(dfs_preorder_nodes(graf, korzeń))
# lub
# list(dfs_postorder_nodes(graf, korzeń))

list(nx.dfs_preorder_nodes(tree, 5))
```

### BFS

```python
# By zastosować algorytm BFS, skorzystać należ z funkcji
# list(bfs_predecessors(graf, korzeń))
# lub
# list(bfs_successors(graf, korzeń))

list(nx.bfs_predecessors(tree, 5))
```
